- seaborn 사용한 시각화
    - matplotlib를 기반으로 한 파이썬 시각화 라이브러리
    - matplotlib의 확장
    - 간결한 문법
    - DataFrame의 데이터를 바로 집계하여 쉽게 시각화 : 그룹화 과정 생략 가능
    
    - [문법] data=데이터프레임, x=x축표시컬럼, y=y축표시컬럼, hue=그룹화 <br><br>
    
- 오차막대 (error bar)
    - 신뢰구간을 의미
    - 0~100까지의 값을 가짐
    - 기본값 95
    - 즉, 추론은 100번 중에 95번 정도 맞다
    - 막대 그래프 위에 표시
    - 부트 스트랩 신뢰구간(Bootstrap confidence interval)
    - 이 데이터를 기반으로 유사한 상황의 95%가 이 범위 내에서 결과를 얻음
    - 신뢰구간은 mean(평균), median(중앙값), mode(최빈값) 등을 고려하여 계산
    - 설정 가능한 값 : ci, pi, se, sd, None <br><br>
    
- estimator
    - np.mean (기본값)
    - np.max
    - np.min
    - np.median
    - np.sum
    - np.count_nonzero :0이 아닌 데이터의 개수

## [가설 2] 코로나 펜데믹 기간(2020~2022)에 서울과 경기의 평당분양가 상승률이 낮았을 것이다.

```python
# Data
s_k = g_local_year.loc[['서울', '경기']]
s_k
```

```python
# 전년 대비 증감률(상승/하락)
# (금년 - 작년) / 작년 * 100
for year in s_k:
    s_k[f'{year}-{year+1}'] = (s_k[year + 1] - s_k[year]) / s_k[year] * 100
    
    if year == 2021:
        break
s_k
```

```python
# 시각화
plt.rc('font', family='D2coding')
s_k.iloc[:, 8:].T.plot(figsize=(10, 6))
```

```python
s_k.iloc[:, 8:].T.plot(kind='bar',
                       figsize=(10, 6),
                       rot=0
                      )
plt.grid(True, axis='y',color='#eee', alpha=0.8, linestyle='--')
plt.legend(loc=(0.03, 0.9), ncols=2, fontsize=10, shadow=True)
```

- [가설 2] 코로나 펜데믹 기간(2020~2022)에 서울과 경기의 평당분양가 상승률이 낮았을 것이다.
    - 코로나 펜데믹 기간 내 서울의 평당분양가는 전년 대비 2020년에는 5.29% 상승했으나, 2021년은 -11.21% 하락하였다. 그러나 2022년에는 전년 대비 14.79% 상승하였다.
    - 경기는 전년 대비 2020년과 2021년 2년간 동일하게 -3.57% 감소하였으나, 2022년은 2.77% 상승하였다.
