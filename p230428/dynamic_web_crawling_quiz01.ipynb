```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests

driver = webdriver.Chrome()

url = 'https://naver.com/'
driver.get(url)

sleep(2)

news_tab = '//*[@id="NM_FAVORITE"]/div[1]/ul[2]/li[2]/a'
news_tab_click = driver.find_element(By.XPATH, news_tab).click()

sleep(2)

economic_tab = '//*[@class="Nlnb_menu_list"]/li[3]/a'
economic_tab_click = driver.find_element(By.XPATH, economic_tab).click()

sleep(2)
```
