# 데이터 인코딩
## 레이블 인코딩 (Label encoding)
- sklearn

```python
# sklearn
from sklearn.preprocessing import LabelEncoder

items = ['사자', '호랑이', '사자', '강아지', '고양이', '고양이', '사슴', '고양이']

# LabelEncoder 객체 생성
encoder = LabelEncoder()

# fit
encoder.fit(items)

# transform
labels = encoder.transform(items)

labels
```

```python
# 인코딩 클래스
encoder.classes_
```

```python
# 숫자 -> 문자열
encoder.inverse_transform([3, 4, 3, 0, 1, 1, 2, 1])
```

```python
# fit과 transform을 함께
encoder.fit_transform(items)
```

## one-hot encoding
- sklearn 
- pandas

### sklearn

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
items = ['사자', '호랑이', '사자', '강아지', '고양이', '고양이', '사슴', '고양이']

# 1차원 -> 2차원
items = np.array(items).reshape(-1, 1)
items
```

```python
# 원-핫 인코딩 객체 생성
oh_encoder = OneHotEncoder()

# fit
oh_encoder.fit(items)

# transform
# sparse matrix (희소 행렬)
# : 값 대부분이 0인 행렬
# : 반대는 밀집행렬 (dense matrix)
oh_labels = oh_encoder.transform(items)
oh_labels
```

```python
# 원-핫 인코딩 (희소행렬) 출력
oh_labels.toarray()
```

```python
# 원-핫 인코딩 차원 
oh_labels.shape
```

```python
# 원-핫 인코딩 타입
type(oh_labels)
```

# 피처 스케일링
- StandardScaler : 평균이 0이고, 분산이 1인 정규 분표 형태로 변환
- MinMaxScaler : 데이터 값을 0~1 범위 값으로 변환

## StandardScaler

```python
from sklearn.datasets import load_iris
import pandas as pd

# 붓꽃 데이터 셋 로딩
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df[:2]
```

```python
# feature들의 평균값
iris_df.mean()
```

```python
# feature들의 분산값
iris_df.var()
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
scaler = StandardScaler()

# [방법 1] fit과 transform 개별 실행
# fit
scaler.fit(iris_df)

# transform
iris_scaler = scaler.transform(iris_df)

# [방법 2] fit과 transform 한번에 실행
iris_scaler = scaler.fit_transform(iris_df)

iris_scaler
```

```python
# array -> dataframe
iris_scaler_df = pd.DataFrame(data=iris_scaler, columns=iris.feature_names)
iris_scaler_df[:2]
```

```python
# feature들의 평균값
iris_scaler_df.mean()
```

```python
# feature들의 분산값
iris_scaler_df.var()
```

## MinMaxScaler
- default : 최솟값 0, 최댓값 1

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 객체 생성
scaler = MinMaxScaler()

# [방법 1] fit과 transform 개별 실행
# fit 
scaler.fit(iris_df)

# transform
iris_scaler_min_max = scaler.transform(iris_df)

# [방법 2] fit과 transform 한번에 실행
# iris_scaler_min_max = scaler.fit_transform(iris_df)

iris_scaler_min_max
```

```python
# array -> dataframe
iris_scaler_min_max_df = pd.DataFrame(data=iris_scaler_min_max, columns=iris.feature_names)
iris_scaler_min_max_df[:2]
```

```python
# feature들의 평균값
iris_scaler_min_max_df.mean()
```

```python
# feature들의 분산값
iris_scaler_min_max_df.var()
```

## scaler를 이용한 fit, transform, fit_transform 적용 시 주의사항!

```python
# 학습 데이터 : 0~10
# 테스트 데이터 : 0~5
train_data = np.arange(0, 11).reshape(-1, 1)
test_data = np.arange(0, 6).reshape(-1, 1)

train_data
```

```python
# 훈련 데이터 스케일링
# MinMaxScaler 객체 생성
# default : 최솟값 0, 최댓값 1로 변환
scaler = MinMaxScaler()

# [방법 1] fit과 transform 개별 실행
# fit() 
scaler.fit(train_data)

# transform()
train_scale = scaler.transform(train_data)

# [방법 2] fit과 transform 한번에 실행
train_scale = scaler.fit_transform(train_data)

# scale된 train_data
train_scale.reshape(-1)
```

```python
# 원본 train_data
train_data.reshape(-1)
```

```python
# 잘못된 테스트 데이터 스케일링

# [방법 1]
scaler.fit(test_data)
test_scale = scaler.transform(test_data)

# [방법 2]
# test_scale = scaler.fit_transform(test_data)

test_scale.reshape(-1)
```

```python
# 잘된 테스트 데이터 스케일링

# 1. scaler 객체 생성
scaler = MinMaxScaler()

# 2-1. fit : 학습 데이터 
scaler.fit(train_data)

# 2-2. 학습 데이터 transform
train_scaler = scaler.transform(train_data)

# 3. transform : 테스트 데이터
test_scale = scaler.transform(test_data)

test_scale.reshape(-1)
```
