## 네이버 증권 -> 국내증시 -> 인기검색종목 -> 더보기 -> 모든 정보 가져오기(30개) 
- 모든 값이 결측치인 행은 삭제
- 타입은 알맞게 변경 (% 그대로 나오게)
- 국내증시.csv 파일로 저장 (인덱스 제외)

```python
import pandas as pd
```

```python
# 1. 웹에서 테이블 가져오기
url = 'https://finance.naver.com/sise/lastsearch2.naver'
table = pd.read_html(url, encoding='euc-kr')
table
```

```python
type(table)
```

```python
table[0]
```

```python
table[1]
```

```python
# 데이터프레임 추출
stock = table[1]
stock
```

```python
stock.info()
```

```python
# 모든 값이 결측치인 행 삭제
stock.dropna(how='all', inplace=True)
stock[:2]
```

```python
# 타입 변경
stock = stock.astype({'순위': int, '현재가': int, '전일비': int, '거래량': int, '시가': int, '고가': int, '저가': int})
stock
```

```python
# 인덱스 재설정
stock.reset_index(drop=True, inplace=True)
stock
```

```python
stock.info()
```

```python
type(stock.검색비율[0])
```

```python
# csv 파일로 저장
stock.to_csv('국내증시.csv', index=False, encoding='UTF-8')
print('저장 완료')
```
