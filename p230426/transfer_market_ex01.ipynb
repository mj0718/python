```python
from urllib import request
import requests
import pandas
```

```python
# 1. 웹 페이지 요청
# [방법 1]
url = 'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop'
# 해당 사이트 개발자 도구에서 네트워크 F5눌러서 맨 위에 있는 항목 클릭 후 맨 아래에 있는 user-agent 복사
# 아래 코드로 서버를 속여서 웹 페이지 접근
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
html = requests.get(url, headers=headers)
html.status_code
```

```python
# 2. 파싱
from bs4 import BeautifulSoup

soup = BeautifulSoup(html.text, 'html.parser')
soup
```

```python
# 3. 엘리먼트 추출
player_info = soup.find_all('tr', class_=['odd', 'even'])
player_info
```

```python
number = []
name = []
position = []
age = []
nation = []
team = []
value = []
```

```python
for player in player_info:
    info = player.find_all('td')
    number.append(info[0].text)
    name.append(info[3].text)
    position.append(info[4].text)
    age.append(info[5].text)
    nation.append(info[6].img['title'])
    team.append(info[7].img['title'])
    value.append(info[8].text.strip())
```

```python
print(number)
print(name)
print(position)
print(age)
print(nation)
print(team)
print(value)
```

```python
# DataFrame 생성
df = pandas.DataFrame({
    'number': number,
    'name': name,
    'position': position,
    'age': age,
    'nation': nation,
    'team': team,
    'value': value
})
df
```

```python
# csv 파일 저장
df.to_csv('market25.csv', index=False)
print('저장 완료')
```
