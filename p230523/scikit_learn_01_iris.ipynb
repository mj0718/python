# 사이킷런을 이용한 붓꽃 데이터 품종 예측

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier ## 결정트리분류모델
from sklearn.model_selection import train_test_split
```

## 데이터 세트 로딩

```python
iris = load_iris()
iris
```

```python
iris_label = iris.target
iris_label
```

```python
iris.DESCR
```

```python
import pandas as pd

iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df[:2]
```

```python
iris_df['label'] = iris.target
iris_df[:2]
```

## 데이터 세트 분리
- 학습 데이터, 테스트 데이터 세트

```python
x_train, x_test, y_train, y_test = train_test_split(iris_data, ## feature data
                                                    iris_label, ## label(target) data,
                                                    test_size=0.2,
                                                    random_state=11,
                                                   )
```

```python
y_test.shape
```

## 학습(Train) 수행
- 학습 데이터로 학습 수행

```python
## DecisionTreeClassifier 객체 생성
dt_clf = DecisionTreeClassifier(random_state=11)
```

```python
## 학습 수행
## x_train : 학습용 feature(문제)
## y_train : 학습용 target(정답)
dt_clf.fit(x_train, y_train)
```

## 예측(Predict) 수행
- 테스트 데이터로 예측 수행
- 문제만 넣어주고 답을 예측

```python
pred = dt_clf.predict(x_test)
pred
```

```python
x_test
```

```python
# 실제 label
y_test
```

## 정확도 평가
- 예측 정확도

```python
from sklearn.metrics import accuracy_score

print(f'예측 정확도 : {accuracy_score(y_test, pred): .4f}')
```

# 만약에 학습 데이터로 예측을 한다면?
- 답을 이미 알고 있어서 정확도가 1이 나옴

```python
# 학습 데이터로 예측
wrong_pred = dt_clf.predict(x_train)
wrong_pred
```

```python
y_train
```

```python
# 예측 정확도 평가
# 정확도 : 1(100%) - 잘못된 Test!
print(f'예측 정확도 : {accuracy_score(y_train, wrong_pred): .4f}')
```

# 판다스 - 데이터 분할

```python
import pandas as pd

iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df[:2]
```

```python
# label(target, 결정값) 컬럼 추가
iris_df['label'] = iris.target
iris_df[:2]
```

```python
feature_data_df = iris_df.iloc[:, :-1]
target_data_df = iris_df.iloc[:, -1]
```

```python
feature_data_df[:2]
```

```python
target_data_df[:2]
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature_data_df,
                                                   target_data_df,
                                                   test_size=0.2,
                                                   random_state=11,
                                                   )
```

```python
type(y_train)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# 결정나무 클래스 생성
dt_clf = DecisionTreeClassifier()
# 훈련
dt_clf.fit(x_train, y_train)
# 예측
pred = dt_clf.predict(x_test)
# 평가 : 정확도
print(f'예측 정확도 : {accuracy_score(y_test, pred): .4f}')
```
