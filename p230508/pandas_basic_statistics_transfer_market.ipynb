```python
import pandas as pd
```

```python
# market100.csv 파일을 읽어서 market 변수에 저장
market = pd.read_csv('market100.csv')
market[:2]
```

```python
# 행과 열 정보
rows, columns = market.shape
print(f'행 개수 : {rows}, 열 개수 : {columns}')
```

```python
# 앞에서 5개의 행 추출
market.head()
market[:5]
market.iloc[:5]
market.loc[:4]
```

```python
# 뒤에서 5개의 행 추출
market.tail()
market[-5:]
market.iloc[-5:]
market.loc[95:]
```

```python
# 10~15위 행 추출
market[9:15]
market.iloc[9:15]
market.loc[9:14]
```

```python
# 이름 컬럼만 추출
market.name
market['name']
market.iloc[:, 1]
market.iloc[:,[1]] # 데이터프레임 형태

market.loc[:, 'name']
market.loc[:, ['name']] # 데이터프레임 형태
```

```python
# 이적료가 가장 높은 5명 선수의 이름
market[:5].name
market[:5]['name']

market.name[:5]
market['name'][:5]

market.iloc[:5, 1]

market.loc[:4, 'name']
```

```python
# 하위 1~5위에 해당하는 선수의 이름, 팀, 이적료 추출
market[['name', 'team', 'value']][-5:]
market.iloc[-5:, [1, -2, -1]]
market.loc[95:, ['name', 'team', 'value']]
```

```python
market.info()
```

```python
# [문제 1] 20세 이하인 선수 정보 추출
market[market.age <= 20]
```

```python
# [문제 2] 20세 이하인 선수 수 추출
market[market.age <= 20].age.count()
```

```python
# [문제 3] 22세인 선수의 이름, 나이, 국가 추출
market[market.age == 22][['name', 'age', 'nation']]
```

```python
# 각 나이의 선수 수
market.age.value_counts()
```

```python
# [문제 4] 축구 선수의 평균 나이
market.age.mean()
```

```python
# [문제 5] 축구 선수의 최고 나이
market.age.max()
```

```python
# [문제 6] 축구 선수의 최소 나이
market.age.min()
```

```python
# [문제 7] 평균 나이 이상인 선수의 이름, 나이, 국가 추출
market[market.age >= market.age.mean()][['name', 'age', 'nation']]
```

```python
# [문제 8] 한국 선수 정보 추출
market[market.nation.str.contains('Korea')]
```

```python
# [문제 9] 토트넘 소속인 선수 정보 추출
market[market.team == 'Tottenham Hotspur']
```

```python
# 이적료 평균
market.value.mean()
```

```python
market[market.age == 24].value.mean()
market[market.age == 24].value.mean().round()
market[market.age == 24].value.mean().round(1)
market[market.age == 24].value.mean().round(2)
market[market.age == 24].value.mean().round(3)
market[market.age == 24].value.mean().round(-1)
```

```python
# 그룹화
group_age = market.groupby(market.age)
group_age
```

```python
for idx, group in group_age:
    print('그룹화 기준 : 나이 >>', idx)
    print(group)
    print(type(group))
    print('='*50)
```
