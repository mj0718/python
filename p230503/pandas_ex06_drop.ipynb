## 행 또는 열 삭제
- drop()

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    [
        range(1,21,2),
        range(2,21,2),
        range(10,101,10)
    ], 
    columns=[f'co{r}'for r in range(1,11)]
)
df
```

### 행 삭제
#### 매개변수 : labels

```python
# 첫 번째 행 삭제
# 복사본으로 작업함 (원본에 영향을 주지 않음)
df.drop(0)
```

```python
# 첫 번째, 마지막 행 삭제
df.drop([0,2])
df.drop({0,2})
# df.drop((0,2)) Error!
```

#### 매개변수 : index

```python
# 첫 번째 행 삭제
df.drop(index=0)
```

```python
# 첫 번째, 마지막 행 삭제
df.drop(index=[0, 2])
```

### 열 삭제
#### 매개변수 : labels

```python
# co2 열 삭제
df.drop('co2', axis=1)
```

```python
# co2, co6 열 삭제
df.drop(['co2', 'co6'], axis=1)
```

#### 매개변수 : columns

```python
# co2 열 삭제
df.drop(columns='co2')
```

```python
# co2, co6 열 삭제
df.drop(columns=['co2', 'co6'])
```

### 원본 반영

```python
# co2 삭제
df.drop(columns='co2', inplace=True)
```
