## 행 또는 열 추가

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    [
        range(1,21,2),
        range(2,21,2),
        range(10,101,10)
    ], 
    columns=[f'co{r}'for r in range(1,11)]
)
df
```

### 행 추가
- 마지막 행 뒤에 추가됨
- iloc는 기존에 있는 인덱스에만 추가가 가능하고 loc는 새로 추가 가능

```python
df.loc[3] = [r for r in range(100, 110, 1)]
df
```

```python
df.loc[5] = [r for r in range(200, 210 ,1)]
df
```

```python
# IndexError
df.iloc[6] = [r for r in range(300, 310 ,1)]
df
```

```python
df.iloc[3] = [r for r in range(400, 410 ,1)]
df
```

```python
# Error
df.loc[100] = [500, 501]
df
```

```python
# 빈 행 추가
df.loc[6] = pd.Series()
df
```

```python
df.loc[8] = 7
df
```

```python
df.info()
```

### 열 추가
#### 마지막 위치에 추가
