```python
# 국가별 나이별 선수의 수
group_nation_age = market.groupby(['nation', 'age'])
group_nation_age.size()
```

```python
# 아르헨티나 22세 데이터프레임 출력
# key가 튜플로 되어 있으므로 튜플로 해야 찾아줌 (리스트는 안 됨)
# group_nation_age.get_group(['Argentina', 22])
group_nation_age.get_group(('Argentina', 22))
```
