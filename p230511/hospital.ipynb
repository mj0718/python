## 병원정보서비스 2021.12.csv 파일 읽어오기
    - hospital 변수에 저장

```python
import pandas as pd
```

```python
hospital = pd.read_csv('병원정보서비스 2021.12.csv', encoding='CP949')
hospital[:2]
```

## 지역별 상급종합병원의 수 : 데이터 -> 시각화

```python
region_hospital = hospital[hospital.종별코드명 =='상급종합'].groupby('시도코드명').size().sort_values(ascending=False)
region_hospital
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('font', size=10)

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

region_hospital.plot(
    kind='pie',
    autopct='%.f%%',
    subplots=True,
    ax=ax[0],
    shadow=True,
    colormap='Set3',
)
region_hospital.plot(
    kind='bar',
    subplots=True,
    ax=ax[1],
    xlabel='도시명',
    color='lightblue',
    rot=0,
)
plt.show()
```
