```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
apt_price = pd.read_csv('../p230515/apt_price.csv')
apt_price[:2]
```

```python
plt.rc('font', family='D2coding')
sns.violinplot(data=apt_price, x='연도', y='분양가격').figure.set_size_inches(12, 5)
plt.grid(axis='y', color='#eee')
```

```python
sns.boxplot(data=apt_price, x='연도', y='분양가격')
```

```python
fig = sns.lineplot(data=apt_price, x='연도', y='분양가격', hue='지역명')
fig.figure.set_size_inches(8, 6)
sns.move_legend(obj=fig,
                loc='upper right',
                bbox_to_anchor=(1.2, 1),
               )
```

```python
g_local_year_price_mean = apt_price.groupby(['지역명', '연도']).분양가격.mean().astype(int).unstack()
g_local_year_price_mean
```

```python
len(g_local_year_price_mean.index)
```

```python
g_local_year_price_mean.plot(figsize=(12, 6),
                             xticks=range(0, len(g_local_year_price_mean.index))
                            )
```

```python
## 연도별 전용면적별 분양가격의 평균

plt.figure(figsize=(12, 3))

sns.boxplot(data=apt_price,
            x='연도',
            y='분양가격',
            hue='전용면적',
            #order=[2015, 2016],
            hue_order=['60', '60~85', '85~102', '102~']
           )

plt.legend(ncols=4,
           loc='upper left'
          )
```
