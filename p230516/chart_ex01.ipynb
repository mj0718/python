```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'n1' : [1, 5, 10, 15, 20],
    'n2' : [1, 2, 3, 4, 20], # 여기서 20을 이상치 또는 극단값이라고 함
    'n3' : [1, 17, 18, 19, 20]  # 1이 이상치
})
df
```

```python
# 편향된 값이 크면 평균이 중앙값보다 크고 편향된 값이 작으면 평균이 중앙값보다 작음
df.describe()
```

# kdeplot()
- 커널 밀도 추정 (Kernel Density Estimation)

```python
# KDE
# 커널 밀도 추정 (Kernel Density Estimation)
sns.kdeplot(df.n1)

plt.axvline(x=df.n1.mean(),
            color='r',
            linewidth=2,
            linestyle='--'
           )
plt.axvline(x=df.n1.median(),
            color='g',
            linewidth=2,
            linestyle='--'
           )
```

```python
sns.kdeplot([1, 2, 3, 4, 5])
```

```python
df.n1.hist(bins=3)
```

```python
# [오른쪽 편향] 왜도(skewness)
# Positive-Skewness
# 이상치들의 꼬리가 오른쪽으로 길게 나타남
# 평균값을 오른쪽으로 잡아끔
# 이런 상황에서 평균값이 대부분의 값보다 높음

# 이상치
# 비정상적인 극단값, 극단적인 값
# 이상치로 평균값이 상승 또는 하락된 현상을 편향(skew, 왜도)되었다고 함
# 편향된 데이터와 이상치로 정보가 왜곡되면, 전형적인 값 파악을 위해 중앙값 등 사용

sns.kdeplot(df.n2)
plt.grid()

# 평균 표시
plt.axvline(x=df.n2.mean(),
            color='r',
            linewidth=2,
            linestyle='--',
           )

# 중앙값 표시
plt.axvline(x=df.n2.median(),
            color='g',
            linewidth=2,
            linestyle='--'
           )
```

```python
# 왼쪽 편향 : 이상치가 왼쪽에 있음
sns.kdeplot(df.n3)

plt.axvline(x=df.n3.mean(),
            color='r',
            linewidth=2,
            linestyle='--'
           )
plt.axvline(x=df.n3.median(),
            color='g',
            linewidth=2,
            linestyle='--'
           )
```

```python
sns.kdeplot(df.n1)
sns.kdeplot(df.n2)
sns.kdeplot(df.n3)
```

```python
sns.kdeplot(data=df, x='n1')
```

```python
sns.kdeplot(data=df, y='n1') # 수평
```

```python
sns.kdeplot(x=df.n1,
            fill=True,
            color='skyblue'
           )
```

```python
sns.distplot(df.n1)
```
