```python
'''
json 파일 요청 방식으로 크롤링  (파일을 찾아야함)
네이버 경제 광고 아래 기사 제목 1-10페이지(200개)까지 화면에 출력
-출력 형태 : 1. [뉴스현장] ~~~ 
'''
import requests
import json
import pandas as pd

headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}

news_title = []

for idx in range(1, 11):
    # 1. json 파일 요청
    url = f'https://news.naver.com/main/mainNews.naver?sid1=101&date=%2000:00:00&page={idx}'

    httpResponse = requests.get(url, headers=headers)

    # 2. json 파일 -> dict 변환
    # [방법 1]
    news = httpResponse.json()
    # [방법 2]
    # news_two = json.loads(httpResponse.text)

    # 3. 뉴스 제목 추출
    news_page = json.loads(news['airsResult'])['result']['101']
    news_title.extend(news_page)
    
# 4. 출력 : 뉴스 제목
for idx, news_line in enumerate(news_title, 1):
    print(f'{idx}. {news_line["title"]}')

# 5. 데이터프레임 생성 : 뉴스 정보 저장
table =pd.DataFrame(news_title)

# 6. naver_news.csv 파일 저장 : 뉴스 정보 저장
table.to_csv('naver_news.csv', index=False)
print('==저장 완료==')
```
