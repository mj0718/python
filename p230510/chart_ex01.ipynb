```python
# 선형 그래프
import matplotlib.pyplot as plt
import numpy as np

data = np.arange(0, 10, 0.01)

sin_y = np.sin(data)
cos_y = np.cos(data)

# 한글 지원 설정 : 한글 깨짐 방지
# plt.rc('font', family='Malgun Gothic')
# # 마이너스 깨짐 방지 설정
# plt.rc('axes', unicode_minus=False)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
#################################################

plt.plot(data, sin_y, label='sin')
plt.plot(data, cos_y, label='cos')

# 그래프 제목
plt.title('sin cos 그래프')
# 범례 : label도 써야 함 (순서가 있어서 label하고 legend 써야 함)
plt.legend()
# 격자 무늬
plt.grid()
# x축 이름
plt.xlabel('x축')
# y축 이름
plt.ylabel('y축')

plt.show()
```

```python
# 원형 그래프
import matplotlib.pyplot as plt

# 데이터
data = [10, 20, 30, 5]

# 그래프
plt.pie(data, 
        labels=['하나', '둘', '셋', '넷'],  # labels는 값 표시
        autopct='%.1f%%',   # 비율 표시 (%f로 해서 실수로 나타내줌)
        startangle=45,    # 시작 위치 변경
#         counterclock=False,   # 시계 방향으로 시작
        explode=[0.1, 0, 0, 0],    # 중심에서 벗어난 정도 (데이터 개수만큼 설정해야 함)
        shadow=True,     # 그림자 설정 
#         colors=['#ff9999', '#ffc000', '#8fd9b6', '#d395d0'],     # 색깔 설정
        colors=['gold', 'silver', 'whitesmoke', 'lightgray'],
        wedgeprops={'width':0.8, 'edgecolor':'k', 'linewidth':1},     # 스타일 지정
        # width : 반지름에 대한 비율
        # edgecolor : 조각 테두리 색깔
        # linewidth : 조각 테두리 두께
       ) 

# 차트 실행
plt.show()
```

```python
# 원형 그래프
# 위 코드 정리
import matplotlib.pyplot as plt

# 데이터
data = [10, 20, 30, 5]

labels=['하나', '둘', '셋', '넷']
explode=[0.1, 0, 0, 0]
#colors=['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
colors=['gold', 'silver', 'whitesmoke', 'lightgray']
wedgeprops={'width':0.8, 'edgecolor':'k', 'linewidth':1}
# 그래프
plt.pie(data, 
        labels=labels, 
        autopct='%.1f%%',   
        startangle=45,   
        explode=explode,    
        shadow=True,    
        colors=colors,
        wedgeprops=wedgeprops,   
       ) 

# 차트 실행
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 1000개 생성
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# 히스토그램 1
plt.figure()
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='bar')
plt.title('histtype - bar')

# 히스토그램 2
plt.figure()
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='barstacked')
plt.title('histtype - barstacked')

# 히스토그램 3
plt.figure()
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='step')
plt.title('histtype - step')

# 히스토그램 4
plt.figure()
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='stepfilled')
plt.title('histtype - stepfilled')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 1000개 생성
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# 히스토그램 1
plt.subplot(2, 2, 1)
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='bar')
plt.title('histtype - bar')

# 히스토그램 2
plt.subplot(2, 2, 2)
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='barstacked')
plt.title('histtype - barstacked')

# 히스토그램 3
plt.subplot(2, 2, 3)
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='step')
plt.title('histtype - step')

# 히스토그램 4
plt.subplot(2, 2, 4)
plt.hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='stepfilled')
plt.title('histtype - stepfilled')

plt.tight_layout(pad=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 1000개 생성
data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 1

# fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig, axs = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.5)

# 히스토그램 1
axs[0][0].hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='bar')
axs[0][0].set_title('histtype - bar')

# 히스토그램 2
axs[0][1].hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='barstacked')
axs[0][1].set_title('histtype - barstacked')

# 히스토그램 3
axs[1][0].hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='step')
axs[1][0].set_title('histtype - step')

# 히스토그램 4
axs[1][1].hist([data1, data2], bins=30, label=('data1', 'data2'), histtype='stepfilled')
axs[1][1].set_title('histtype - stepfilled')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# x축
x = np.arange(0, 100)

# y축
y1 = [v*v for v in x]
y2 = [-v for v in x]
y3 = [v*v/2 for v in x]

# 그래프 그리기
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1) # 행, 열, 몇 번째
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x, y1, x, y3)
ax2.plot(x, y2)

# 그래프 사이 간격 조정
plt.tight_layout(pad=2)

# 차트 실행
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# x축
x = np.arange(0, 100)

# y축
y1 = [v*v for v in x]
y2 = [-v for v in x]
y3 = [v*v/2 for v in x]

# 그래프 그리기
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.subplots_adjust(hspace=0.5) # 그래프 사이 간격 조정
fig.set_size_inches(10, 8)   # 그래프 창 크기 설정

# 첫 번째 subplot
ax1.plot(x, y1, x, y3)
ax1.set_xlabel('x축')
ax1.set_ylabel('y축')
ax1.grid()

# 두 번째 subplot
ax2.plot(x, y2)
ax2.set_title('제목')
```

```python
import matplotlib.pyplot as plt

one = [10, 20, 30, 40]
two = [40, 30, 20, 10]

# 제목
plt.title('color')

# 그래프
plt.plot(one,'b.', label='one')  # .은 산점도
plt.plot(two, 'r^', label='two')  # ^는 삼각형

# 범례 표시
plt.legend(loc=2)
```
